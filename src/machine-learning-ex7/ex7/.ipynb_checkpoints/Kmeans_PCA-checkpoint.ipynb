```python
df = loadmat('ex7data2.mat')
df.keys()
```

```python
X = df['X']
```

```python
X.shape
```

```python
plt.scatter(X[:,0], X[:,1], color='g')
```

```python
k =3
max_iters = 10
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters = k, max_iter = max_iters)
```

```python
clf.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], s=60, c=clf.labels_, cmap= plt.cm.RdYlGn)
plt.scatter(clf.cluster_centers_[:,0], clf.cluster_centers_[:,1], marker='+', s=100, c='b', linewidth=2);
```

## Principle Component Analysis

```python
df3 = loadmat('ex7data1.mat');
```

```python
df3.keys()
```

```python
X2 = df3['X']
```

```python
X2.shape
```

```python
plt.scatter(X2[:,0], X2[:,1], s=60)
```

```python
from sklearn.preprocessing import StandardScaler
from scipy import linalg
```

```python
scaler = StandardScaler()
scaler.fit(df3['X'])
```

```python
U, S, V = linalg.svd(scaler.transform(df3['X']).T)
```

```python
print (U), (S), (V)
```

```python
plt.scatter(df3['X'][:,0], df3['X'][:,1], s=30, edgecolors='b',facecolors='None', linewidth=1);
# setting aspect ratio to 'equal' in order to show orthogonality of principal components in the plot
plt.gca().set_aspect('equal')
plt.quiver(scaler.mean_[0], scaler.mean_[1], U[0,0], U[0,1], scale=S[1], color='r')
plt.quiver(scaler.mean_[0], scaler.mean_[1], U[1,0], U[1,1], scale=S[0], color='r');
```
